```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import re
```

```python
response = requests.get('https://www.armp-rdc.org/index.php/fr/prestations-itellectuelles?limit=0&start=0')
offres = {}

html = response.content
soup = bs(html, "html.parser")
table = soup.find("table", class_="category")
tbody = table.find("tbody")
a = tbody.find_all("a")
date_list = tbody.find_all("td",class_="list-date")
offres['titre'] = [_.text.strip() for _ in a]
offres['date'] = [_.text.strip() for _ in date_list]
```

```python
offres_df = pd.DataFrame(offres)
offres_df.head()
```

```python
offres_df.to_csv('offres.csv', index=False)
```
